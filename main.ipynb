```python
import geocoder

import geopandas as gpd

g = geocoder.ip('me')

latlng = g.latlng
print(latlng)
convertedlat = str(latlng[1])
convertlong = str(latlng[0])
world = gpd.read_file( gpd.datasets.get_path('naturalearth_lowres') )
ax = world.plot()
ax.plot(latlng[1], latlng[0],  marker= ".", color="green" )

print("Your cordinates are " + convertedlat + ',' + convertlong)
```
